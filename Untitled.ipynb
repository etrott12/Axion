# 2. Matter Power Spectrum

### Two-Point Correlation Function
"Given a random galaxy in a location, the correlation function describes the probability that another galaxy will be found within a given distance." -- Peebles

$\xi(\vec{r}) = \langle \rho(\vec{r})\rho(0) \rangle$

The matter power spectrum is the Fourier Transform of the two-point correlation function.

$P(\vec{k}) = \frac{1}{(2\pi)^{3/2}} \int \xi(\vec{r}) e^{i\vec{k}\cdot\vec{r}} d^3k$

# 3. Angular Power Spectrum

Galaxy surveys often gather photometric redshift data that places galaxies into redshift bins rather than observing each galaxy's specific redshift spectroscopically. We can minimize the effect of this redshift uncertainty by integrating over the line of sight to convert a 3D $P(k)$ into a 2D $P(\ell)$.

This conversion is given by

$P_{ij}(\ell) = \frac{2\pi^2}{c\ell^3} \int_0^\infty H(z) D_A(z) W_i(z) W_j(z) \Delta^2(k;z) dz$

The first two integrand factors are simple redshift evolutions

$H(z) = H_0\sqrt{\Omega_m(1+z)^3+\Omega_\Lambda}$

$D_A(z) = \frac{c}{H_0} \int_0^z \frac{dz}{\sqrt{\Omega_m(1+z)^3+\Omega_\Lambda}}$


```python
# angular diameter distance redshift evolution
def d_a(z):
    dalist = []
    for i in range(len(z)):
        ztest = np.linspace(0.,z[i],z[i]/dz)
        integral = (c/H0)*np.trapz((omegam*(1.+ztest)**3+omegal)**(-0.5),ztest,dz)
        dalist.append(np.float_(integral/(1.+ztest[-1:])))
    dalist[0] = 0.
    return dalist
da = d_a(z)
plt.plot(z,da,color = 'b')
plt.xlabel(r'$z$',size = 12)
plt.ylabel(r'$D_A(z) [\rm{Mpc}]$',size = 12)
plt.title(r'Angular Diameter Distance',size = 12)
plt.show()
plt.close('all')

# hubble constant redshift evolution
def h(z):
    hz = H0*np.sqrt(omegam*(1.+z)**3.+omegal)
    return hz
hz = h(z)
plt.plot(z,hz,color = 'b')
plt.xlabel(r'$z$',size = 12)
plt.ylabel(r'$H(z) [km/s/\rm{Mpc}]$',size = 12)
plt.title(r'Hubble Constant',size = 12)
plt.show()
plt.close('all')
```

The third and fourth terms are window functions. These window functions represent number counts of galaxies in different redshift bins. Due to the redshift uncertainty in photometric surveys, galaxies are probabilistically assigned to a redshift range. Window functions are computed using the prescription

$W_i(z) = b(z)n_i(z)/\bar{n}_i$

where the first factor is the linear galaxy bias

$b(z) = 1+0.84z$

and the second and third terms represent the normalized galaxy number density.

```python
def bins(a,c,zmax):
    bins = []
    while a < zmax:
        b = (a*(2.+c)+(2.*c))/(2.-c)
        bins.append([a,b])
        a = b
    return bins

# zlist = bins(0.15,0.04547,3.5)
zlist = [[0.0,0.5],[0.5,1.0],[1.0,1.5],[1.5,2.0],[2.0,2.5],[2.5,3.0],[3.0,3.5]]
zavg = [np.average(zlist[i]) for i in range(len(zlist))]


def window(z,i,use_n1 = True):
    a = zlist[i][0]
    b = zlist[i][1]
    bias = 1.+(0.84*z)
    def error(x,z):
        error = sp.special.erf(10.*np.sqrt(2.)*(x-z)/(1.+z))
        return error
    def Iratio(a,b,z):
        Iratio = (error(b,z)-error(a,z))/(1.+error(2.*z,z))
        return Iratio
    def ni2(i):
        xi = (zavg[i]-z)/(np.sqrt(2.)*sigmaz)
        xi1 = (zavg[i+1]-z)/(np.sqrt(2.)*sigmaz)
        ni2 = 0.5*A2*n2*(sp.special.erf(xi1)-sp.special.erf(xi))
        return ni2
    if use_n1 == True:
        n = A*np.square(z)*np.exp(-2.*z)
        ni = Iratio(a,b,z)*n
        nibar = np.trapz(ni,z,dz)
        window = bias*ni/nibar
    else:
        n2 = np.square(z)*np.exp(-np.square(z/0.92))
        ni2 = ni2(i)
        nibar2 = np.trapz(ni2,z,dz)
        window = bias*ni2/nibar2
    return window

wi = []
wi2 = []
for i in range(len(zlist)-1):
    wi.append(window(z,i,use_n1 = False))
    wi2.append(window(z,i,use_n1 = True))

for i in range(len(wi)):
    plt.plot(z,wi[i],color = 'b')
    plt.plot(z,wi2[i],color = 'r')
plt.show()

test = [np.trapz(wi[i],z,dx=dz) for i in range(len(wi))]
print(test)
```
